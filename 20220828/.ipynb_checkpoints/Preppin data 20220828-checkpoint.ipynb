Input data
 - Merge Km's and min's as Minutes
 - Change 'Value' to Mins
 - Split up the unnamed column into:
     - Coach
     - Calories
     - Music Type
 - Change Music Type values to be Title Case (first letter of each word is capitalised)
 - Create three parameters:
 - Music Type
 - Coach
 - Top N 
 - Create a way to return the Top N value selected and order the file with the highest calories burnt at the top
 - Create filters so only the parameter selection remains in the output data set
 - For Top N parameter it's all the values up to that number
 - Output the data but use the parameter values in the file name so the CEO knows what it contains

```python
import os
import pandas as pd
```

```python
df = pd.read_csv('Preppin data 20220828.csv')
```

```python
df.columns = [i.strip().lower().replace(' ','_').replace(':','') for i in df.columns]
```

```python
df.units = 'min'
df = df.rename(columns={'value':'mins'})
```

```python
df[['coach','calories','music_type']] = df.unnamed_4.str.split('-',expand=True)
```

```python
df.drop(columns=['unnamed_4','units'],inplace=True)
```

```python
df.music_type=df.music_type.str.title()
```

```python
# def make_selection():
activate = 1
while activate == 1:
    selections = {1: 'Music Type',
                  2: 'Coach',
                  3: 'Exit'}
    print('Filter and sort by:')
    for k,v in selections.items():
        print(k,':',v)

    user_selection = int(input())
    
    if user_selection not in list(selections.keys()):
        print('invalid selection, please try again')
   
    elif user_selection in list(selections.keys())[:-1]:
        param_selection = selections[user_selection].lower().replace(' ','_')
        not_selected = [i for i in list(selections.values()) if i not in [selections[user_selection],'Exit']][0].lower().replace(' ','_')
        

        activate = 0
    elif user_selection in list(selections.keys())[-1:]:
        print('you have exited the program')
        activate = 0

top_n = input('Return the top n values: ')

print(f'You have chosen to show the top {top_n} by {param_selection}.')
```

```python
df.loc[df.coach =='Kym ',[param_selection,'calories',not_selected,'date','mins']].sort_values('calories',ascending = False)
```
