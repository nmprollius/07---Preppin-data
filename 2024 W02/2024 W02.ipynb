```python
import os
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv(r'C:\Users\nprol195\Documents\02 - Weekly\07 - Preppin data\2024 W01\With Cards.csv')
df2=pd.read_csv(r'C:\Users\nprol195\Documents\02 - Weekly\07 - Preppin data\2024 W01\Without Cards.csv')
```

```python
df=pd.concat([df1,df2])
del df1,df2
```

```python
from datetime import date as dt
df['date']=pd.to_datetime(df['date']).dt.quarter
df.rename(columns={'date':'quarter'},inplace=True)
```

```python
df=df.groupby(['quarter','flow_card','class']).describe()['price'].unstack(level='class').reset_index().loc[:,['flow_card','quarter','mean']]
```
