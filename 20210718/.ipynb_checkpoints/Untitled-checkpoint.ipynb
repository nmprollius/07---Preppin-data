```python
dta = pd.read_excel('InternationalPenalties.xlsx', sheet_name=None)
```

### Determine what competition each penalty was taken in

```python
for i in dta.keys():
    dta[i]['competition'] = i
    dta[i].columns = [clm.lower().strip().replace(' ','_') for clm in dta[i].columns]
 

df = pd.DataFrame()
for i in dta.keys():
    df=df.append(dta[i]).reset_index(drop=True)
    
# df
# df.to_csv('combined.csv')
```

### Clean any fields, correctly format the date the penalty was taken, & group the two German countries (eg, West Germany & Germany)

```python
# df.info()
```

```python
df['event_year']=df['event_year'].apply(lambda x: x.replace(',',''))
df['winner']=df['winner'].astype('str')
df['winner']=df['winner'].apply(lambda x : x.replace('\xa0',''))
df['loser']=df['loser'].apply(lambda x : x.replace('\xa0',''))
df['winning_team_taker']=df['winning_team_taker'].astype('str')
df['winning_team_taker']=df['winning_team_taker'].apply(lambda x : x.replace('\xa0','_'))
df['losing_team_taker']=df['losing_team_taker'].astype('str')
df['losing_team_taker']=df['losing_team_taker'].apply(lambda x : x.replace('\xa0','_'))
df.replace(to_replace={'West Germany':'Germany'},inplace=True)
```

### Rank the countries on the following: 
Shootout win % (exclude teams who have never won a shootout)

Penalties scored %

### Shootout win % (exclude teams who have never won a shootout)

```python
df['winning_team_scores'] =  df['winning_team_taker'].apply(lambda x: 1 if 'scored' in x else 0)
df['losing_team_scores'] =  df['losing_team_taker'].apply(lambda x: 1 if 'scored' in x else 0)
```

```python
df2 = df[['winner','loser', 'date', 'competition', 'winning_team_scores', 'losing_team_scores']].copy()
df2 = df2.groupby(['winner', 'loser', 'date', 'competition']).sum().reset_index().copy()
df2['no_records'] = 1
```

```python
win = df2[['winner','no_records']].groupby('winner').sum().reset_index()
win.rename(columns={'no_records':'wins','winner':'team'},inplace=True)
lose = df2[['loser','no_records']].groupby('loser').sum().reset_index()
lose.rename(columns={'no_records':'losses','loser':'team'},inplace=True)
```

```python
shootouts = win.append(lose).fillna(0) 
shootouts = shootouts.groupby('team').sum()
shootouts['total_games'] = shootouts['wins']+shootouts['losses']
shootouts['pct'] =  round(shootouts['wins']/shootouts['total_games']*100)
shootouts=shootouts[shootouts['wins'] > 0].sort_values(['pct','wins'],ascending=False).reset_index().copy()
shootouts['rank'] = shootouts['pct'].rank(method='dense',ascending=False)
shootouts
```

### Penalties scored %

```python
df3 = df.copy()
df3 = df3[df3['winning_team_taker'] != 'nan']
df3['tf']=df3['winning_team_taker'].apply(lambda x: 'scored' in x)
df3['penalties_scored'] = df3['tf'].apply(lambda x: 1 if x else 0)
df3['penalties_missed'] = df3['tf'].apply(lambda x: 1 if not x else 0)
win = df3[['winner','penalties_scored', 'penalties_missed']].copy()
win['total'] = win['penalties_scored']+win['penalties_missed']
win = win.groupby('winner').sum().copy()
```

```python
df3 = df.copy()
df3 = df3[df3['losing_team_taker'] != 'nan']
df3['tf']=df3['losing_team_taker'].apply(lambda x: 'scored' in x)
df3['penalties_scored'] = df3['tf'].apply(lambda x: 1 if x else 0)
df3['penalties_missed'] = df3['tf'].apply(lambda x: 1 if not x else 0)
lose = df3[['loser','penalties_scored', 'penalties_missed']].copy()
lose['total'] = lose['penalties_scored']+lose['penalties_missed']
lose = lose.groupby('loser').sum().copy()
```

```python
penalties_pct = win.append(lose).reset_index().groupby('index').sum()
penalties_pct['pct'] = round(penalties_pct['penalties_scored']/penalties_pct['total']*100)
penalties_pct.sort_values(['pct','penalties_scored'],ascending=False)
```

### What is the most and least successful time to take a penalty? (What penalty number are you most likely to score or miss?)

```python
df.columns
```

```python
df[['no.', 'penalty_number', 'event_year', 'winner', 'full_time_score',
       'loser', 'winning_team_gk', 'winning_team_taker', 'losing_team_taker',
       'losing_team_gk', 'round', 'date', 'competition', 'winning_team_scores',
       'losing_team_scores']]
```

```python
windf=df[['penalty_number', 'winner', 'winning_team_taker', 'winning_team_scores']].copy()
```

```python
losedf = df[[ 'penalty_number','loser', 'losing_team_taker', 'losing_team_scores']].copy()
```

```python
for i in [windf,losedf]:
    i.columns = [ 'penalty_number','team', 'team_taker', 'score']
teamdf = windf.append(losedf)
teamdf['misses']=teamdf['score'].apply(lambda x: 1 if x == 0 else 0)    
teamdf = teamdf.groupby(['penalty_number']).sum().copy()
```

```python
teamdf['total'] = teamdf['score']+teamdf['misses']
teamdf['pct'] = round(teamdf['score']/teamdf['total']*100)
teamdf['rank'] = teamdf['pct'].rank(method='dense',ascending=False)
teamdf.sort_values('pct',ascending=False)
```
