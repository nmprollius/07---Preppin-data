```python
# Imports
import os
import pandas as pd
import numpy as np
import datetime as dt
import shutil
```

```python
# Download sheet

# shutil.copy(r'c:\Users\nprol195\Downloads\Strange table structure.xlsx',os.getcwd()+'\\Strange table structure.xlsx')

df = pd.read_excel('Strange table structure.xlsx')
```

```python
yr = df.columns[0].split(' ')[-1]
```

```python
# Get column names from multiple rows
level1 = df.loc[2]#.reset_index(drop=True)
level2 = df.loc[1]#.dropna().reset_index(drop=True)

clm_names = list(level1 + '_' + level2.ffill())
clm_names[0] = 'store'
df.columns = clm_names
```

```python
# Filter nan rows
df = df.loc[(~df.store.isna()) & (df.store !='Store')]
```

```python
vars =  ['store']
vals = [i for i in df.columns if i not in vars]
df = df.melt(id_vars=vars, value_vars=vals)
```

```python
# transform date
df[['var','month']] = df.variable.str.split('_',expand=True)
df.month = df.month + yr
df.month = pd.to_datetime(df.month,format='%b%Y')
```

```python
df = df.drop(columns='variable').pivot_table(  values='var',
        index=['store','month'],
        columns='var',
        aggfunc="sum").reset_index()
```
