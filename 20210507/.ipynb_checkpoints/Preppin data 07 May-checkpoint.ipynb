```python
import os
import pandas as pd
import datetime as dt
import numpy as np
```

```python
df = pd.read_excel('PD 2021 Wk 18 Input.xlsx')
```

```python
df.head()
```

```python
df.columns = [x.lower().strip().replace(' ','_') for x in df.columns]
```

```python
temp = df['completed_in_days_from_scheduled_date'].apply(lambda x: pd.Timedelta(x, unit='D'))
temp
```

```python
df['completed_date'] = df['scheduled_date'] + temp
```

```python
df = df.rename(columns={'completed_in_days_from_scheduled_date':'days_difference_to_schedule'}).copy()
```

```python
df2 = df[['project', 'sub-project', 'task', 'owner', 'completed_date']].set_index(['project', 'sub-project', 'owner','task']).unstack().reset_index().copy()
```

```python
df2
```

```python
df2['scope_to_build'] = df2['completed_date']['Build']-df2['completed_date']['Scope']
df2['build_to_delivery'] = df2['completed_date']['Deliver']-df2['completed_date']['Build']
```

```python
df2final = df2.set_index(['project','sub-project','owner','scope_to_build','build_to_delivery']).stack().copy()
```

```python
df2final = df2final.reset_index()
```

```python
mergeddf = df.merge(df2final,how='outer',on=['project', 'sub-project', 'owner','task','completed_date'])
```

```python
mergeddf['weekday'] = mergeddf['completed_date'].apply(lambda x:x.date().strftime('%A'))
```

```python
mergeddf[['weekday','task','build_to_delivery','scope_to_build',
          'days_difference_to_schedule','project',
          'sub-project','owner','scheduled_date','completed_date']].set_index('weekday')
```
