```python
import os
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv('Preppin data 20220706.csv')
df.columns= [i.lower().strip().replace(' ','_') for i in df.columns]
df['sale_date']= pd.to_datetime(df['sale_date'])
```

```python
df['sale_date']=df['sale_date'].dt.date
df[['product','quantity']] = df['product_name'].str.split('-',expand=True)
df['product']=df['product'].str.strip()
clean_map = {' 25ml' : 25,
                ' 50ml' : 50,
                ' 100ml' : 100,
                ' 250ml' : 250,
                ' 500ml' : 500,
                ' 750ml' : 750,
                ' 1L' : 1000,
                ' 1x' : 1,
                ' 2x' : 2,
                ' 4x' : 4
            }
df['quantity']=df['quantity'].replace(clean_map)

df['order_id']=df['order_id'].astype('category')
# df['quantity']=df['quantity'].astype('category')
```

```python
cols_to_use = ['order_id', 'sale_value', 'store_name',
                'region', 'quantity']

liquid_df = df[df['product']=='Liquid'].reset_index(drop=True)[cols_to_use].copy()
bar_df = df[df['product']=='Bar'].reset_index(drop=True)[cols_to_use].copy()
```

```python
liquid_df.groupby(['quantity','store_name', 'region']).sum().sort_values(['region','sale_value']).reset_index()
```
